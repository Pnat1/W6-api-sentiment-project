To introduce data in column "line" we make use of some defensinve programming in the form of created function checking tools that will always make sure data is not repeated or missing.

```python
def check(elem,string):
    """
    Parametrized function that checks for an existing actor, place or line.
    Return True if they do and False if they do not
    """
    if elem == "actors":
        query = list(engine.execute(f"SELECT film_character FROM actors WHERE film_character = '{string}'"))
        if len(query) > 0:
            return True
        else:
            return False
        
    elif elem == "place":
        query = list(engine.execute(f"SELECT place FROM place WHERE place = '{string}'"))
        if len(query) > 0:
            return True
        else:
            return False
        
    elif elem == "line_quotes":
        query = list(engine.execute(f"SELECT line_quote FROM line_quotes WHERE line_quote = '{string}'"))
        if len(query) > 0:
            return True
        else:
            return False
```

In order to complete the line_quotes sql table, defensive programming is used to check wether values exist. I will include the condition that if they do not exist then a value must replace them with an insert query.

```python
def insertplace(string):
    """
    Calls check function to see if place exists
    Inserts place if it does not exist
    """
    if check("place", string):
        return "place exists"
    else:
        engine.execute(f"INSERT INTO place (place) VALUES ('{string}');")
```

```python
insertplace("int. Chevy Nova")
```

```python
def insertactor(name):
    """
    Calls check function to see if film_character exists
    Inserts film_character if it does not
    """
    if check("actors", name):
        return "film character exists"
    else:
        engine.execute(f"INSERT INTO actors (film_character) VALUES ('{name}');")
```

```python
insertactor("Marsellus")
```

We need the ID's of both film characters and place identified so that we can insert them into the sql line_quotes table

```python
def giveId(elem,string):
    """
    Gives back the ID of what we ask knowing the element exists
    """
    if elem == "character":
        return list(engine.execute(f"SELECT actor_id FROM actors WHERE film_character ='{string}';"))[0][0]
    elif elem == "place":
        return list(engine.execute(f"SELECT place_id FROM place WHERE place ='{string}';"))[0][0]
```
