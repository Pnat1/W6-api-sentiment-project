```python
df1 = df.rename(columns= {"Line number":"line_number", "Character (in script)": "character_script", "Character (actual)": "character", "Off screen": "off_screen", "Voice-over": "voice_over", "Place": "place", "Time": "time", "Line": "line", "Word count": "word_count"})
df1.head()
```

The CSV file contains each line of dialogue as a row, containing the following columns:

- line_number	

- character_script:
Some characters have multiple names in the script, such as "Young man" before we learn the actual name later on. It might be insteresting to analyse their sentiment in different stages of the film before and after they change names.

- character:
See "character_script"; this is the actual character name, even if the script has different name written for this piece of dialogue.

- off_screen:
Whether the line is said off screen or not: the character is in the scene location but not seen on camera.

- voice_over:
Whether the line is voiced over or not: the character is not in the scene location, but we hear their voice.

- place:
The scene location where line is said

- time:
The time of day when the line is said: either "morning", "day", or "night"

- line:
The line itself

- word_count:
Number of words in the line.

I drop all the columns I am not going to use

```python
df2 = df1.drop(["line_number", "character_script", "off_screen", "voice_over", "time", "word_count"], axis= 1)
df2
```
